```python
#import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjust the confirmation of the plots

#Read in the data
df= pd.read_csv(r'C:\Users\cokpo\Documents\Projects\Python\movies.csv')
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)

df
```

```python
#Possiblitites for the dataset:
#Budget high correlation(The more money they spent, the more they bring in)
#Company High correaltion (bigger companies bring in more money)

#compare using scatterplot
```

```python
#Scatterplot with Budget vs Gross

plt.scatter(x=df['budget'],y=df['gross'])

plt.title("Budget vs Gross Earnings")

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnings')

plt.show
```

```python
#Regression plot to identify the correlation
#Plot Budget vs Gross using seaborn

sns.regplot(x='budget', y = 'gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
df.corr(method = 'spearman')
```

```python
#There is a high correlation between Budget and Gross. Observation confirms prediction.
```

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()


#The brighter the colour, the higher the correlation and vice versa
```

```python
#correlation between company and gross
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs  = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
